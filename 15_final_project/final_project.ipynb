# Финальный проект. Телеком

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

### Описание услуг

Оператор предоставляет два основных типа услуг: 

1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL*,* от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).  

Также доступны такие услуги:

- Интернет-безопасность: антивирус (*DeviceProtection*) и блокировка небезопасных сайтов (*OnlineSecurity*);
- Выделенная линия технической поддержки (*TechSupport*);
- Облачное хранилище файлов для резервного копирования данных (*OnlineBackup*);
- Стриминговое телевидение (*StreamingTV*) и каталог фильмов (*StreamingMovies*).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

### Описание данных

Данные состоят из файлов, полученных из разных источников:

- `contract.csv` — информация о договоре;
- `personal.csv` — персональные данные клиента;
- `internet.csv` — информация об интернет-услугах;
- `phone.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

## Предобработка данных

### Унификация названий столбцов

Приведём названия столбцов к единому стилю CamelCase.

```python
for desc, df in df_dict.items():
    print('Таблица', desc)
    
    df_cols = df.columns.to_list()
    print('До:\t', ', '.join(df_cols))
    
    for i, col in enumerate(df_cols):
        if col[0].islower():
            df_cols[i] = col[0].upper() + col[1:]
            
    df.columns = df_cols
    print('После:\t', ', '.join(df.columns))
    print()
```

### Приведение типов

Для начала объединим все таблицы в один датафрейм, посчитав перед этим, сколько полных записей клиентов останется в полученном датафрейме. В любом случае таких записей будет не больше числа всех записей в таблице **internet**.

```python
for desc, df in df_dict.items():
    if df is not contract_df:
        inconsistencies_num = (~df['CustomerID'].isin(contract_df['CustomerID'])).sum()
        print('В датафрейме "contracts" отсутствует {} записей клиентов из датафрейма "{}"'.format(inconsistencies_num, desc))
```

Таблицы **contract** и **personal** имеют одинаковое количество записей, значит, различие в таблицах **internet** и **phone** приведёт нас к числу полных записей в итоговом датафрейме.

```python
inconsistencies_num = len(contract_df['CustomerID']) - len(internet_df['CustomerID'])
print('В датафрейме "internet" отсутствует {} записей клиентов из датафрейма "contract"'.format(inconsistencies_num))

inconsistencies_num = (~internet_df['CustomerID'].isin(phone_df['CustomerID'])).sum()
print('В датафрейме "phone" отсутствует {} записей клиентов из датафрейма "internet"'.format(inconsistencies_num))
```

Значит, в 1526 записях будут пропущены значения из услуг интернета таблицы **internet**, а ещё в 682 записях будет пропущено значение в столбце *'MultipleLines'*.

Объединим таблицы по столбцу *'CustomerID'* методом *outer* и заполним пропуски в булевых признаках 'No', а в категориальных - наиболее популярным признаком во избежание ощутимой потери информации.

```python
buf = None
for df in df_dict.values():
    if buf is None:
        buf = df
    else:
        buf = buf.merge(df, on='CustomerID', how='outer')
        
buf['InternetService'].fillna(value=buf['InternetService'].mode().values[0], inplace=True)
customer_full_df = buf.fillna('No')
customer_full_df.info()
```

Приведём к типу дат дату заключения договора.

```python
customer_full_df['BeginDate'] = pd.to_datetime(customer_full_df['BeginDate'])
```

Добавим в датафрейм целевой столбец на основе даты расторжения: если есть дата, то клиент расторг договор с оператором связи.  
Посмотрим на соотношение между двумя классами.

```python
customer_full_df['IsQuit'] = (customer_full_df['EndDate']!='No').apply(int)
customer_full_df['IsQuit'].mean()
```

Действующих клиентов в датафрейме немногим более чем в два раза больше ушедших.

Заменим 'No' в столбце даты расторжения договора на дату выгрузки данных и приведём к типу дат.

```python
customer_full_df['EndDate'] = pd.to_datetime(customer_full_df['EndDate'].replace('No', '2020-02-01'))
```

Заменим классы 'Yes'/'No' во всех признаках на 1/0.

```python
def yes_no_to_bool(x):
    if x == 'Yes':
        return 1
    elif x == 'No':
        return 0
    else:
        return x

for col in customer_full_df.columns:
    unique_vals = customer_full_df[col].unique()
    if 'Yes' in unique_vals:
        customer_full_df[col] = customer_full_df[col].apply(yes_no_to_bool)

customer_full_df.info()
```

Приведём к вещественному типу значения в столбце суммарных платежей, неявно пропущенные значения заменим на NaN.  
Выведем количество пропущенных значений.

```python
customer_full_df['TotalCharges'] = pd.to_numeric(customer_full_df['TotalCharges'], errors='coerce')
customer_full_df['TotalCharges'].isnull().sum()
```

Посмотрим, что это за клиенты.

```python
customer_full_df[customer_full_df['TotalCharges'].isnull()]
```

Видно, что данные клиенты заключили договор в феврале и скорее всего не успели ещё оплатить услуги, поэтому можно заполнить пропуски нулями.

```python
customer_full_df['TotalCharges'].fillna(0, inplace=True)
customer_full_df.isnull().sum()
```

Построим диаграмму размаха для суммарных платежей.

```python
plt.figure(figsize=(10,3))
sns.violinplot(customer_full_df['TotalCharges'])
plt.title('Распределение суммарных платежей');

customer_full_df['TotalCharges'].describe()
```

Распределение скорее всего унимодально с перекосом в сторону 0, также нет выбросов.

Коэффициент корреляции Пирсона не врёт: между указанными столбцами действительно близкая к линейной связь.

Теперь проверим наличие связей между категориальными признаками. Здесь применим статистику на основе теста $X^2$ - статистику Крамера (V) с коррекцией Бергсмы-Уичера для уменьшения стремления статистики переоценивать силу связи. Она возвращает силу связи между двумя категориальными признаками в диапазоне от 0 до 1, где 0 - зависимость отсутствует, 1 - максимальная зависимость.

```python
def cramers_v(confusion_matrix):
    """
    Вычисляет статистику Крамера (V) для категориально-категориальной связи.
    Использует коррекцию от Бергсмы и Уичера.
    Журнал Корейского Статистического Общества 42 (2013): 323-328.
    
    Аргументы
    ---------
    confusion_matrix : array_like
        Таблица сопряжённости
        
    Возвращает
    ----------
    Значение статистики Крамера (V).
    
    """
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))


corr_coefs = []
boolean_features = [
    col for col in customer_prepared_df.columns
    if customer_prepared_df[col].dtype == 'int64' and col not in numeric_features
]
categorial_features_all = categorial_features + boolean_features

for col_i in categorial_features_all:
    corr_coefs_i = []
    
    for col_j in categorial_features_all:
        
        confusion_matrix = pd.crosstab(index=customer_prepared_df[col_i], columns=customer_prepared_df[col_j])
        corr_coef = cramers_v(confusion_matrix.values)
        
        corr_coefs_i.append(corr_coef)
        
    if corr_coefs_i:
        corr_coefs.append(corr_coefs_i)
    
np.array(corr_coefs)
```

Построим матрицу корреляций на основе посчитанных значений статистики Крамера (V).

```python
corr_matrix = pd.DataFrame(corr_coefs, index=categorial_features_all, columns=categorial_features_all)

plt.figure(figsize=(12, 12))
sns.heatmap(corr_matrix, robust=True, square=True, annot=True)

plt.title('Корреляционная матрица категориальных данных')
plt.show()
```

Как видим, максимальная сила связи 0.53 между принаками *'StreamingMovies'* и *'StreamingTV'*, что не является критичным.

Далее оценим силу связи между категориальными и количественными признаками с помощью одностороннего анализа дисперсий (**oneway ANOVA**). Возвращаемое значение F-статистики говорит о силе связи, а значимость влияет на принятие или отвержение нулевой гипотезы, которая в данном случае звучит так: "Выборки во всех группах взяты из популяций с одинаковыми средними значениями". Если средние значения групп взяты из популяций с одинаковыми средними значениями, то дисперсия между средними значениями групп должна быть ниже дисперсии выборок, следуя центральной предельной теореме. Следовательно, более высокое соотношение означает, что выборки были взяты из популяций с различными средними значениями, что говорит о зависимости между категориальным и количественным признаком. Однако о статистически значимом различии средних можно говорить только в случае p-value < выбранного уровня значимости (здесь - 0.5).

```python
fvalues = []
pvalues = []

for cat_col in categorial_features_all:
    fvalues_i = []
    pvalues_i = []
    for num_col in numeric_features:
        series_list = [
            customer_prepared_df[customer_prepared_df[cat_col] == value][num_col]
            for value in customer_prepared_df[cat_col].unique()
        ]
        
        F, pvalue = f_oneway(*series_list)
        
        fvalues_i.append(F)
        pvalues_i.append(pvalue)
        
    fvalues.append(fvalues_i)
    pvalues.append(pvalues_i)
    
fvalues_df = pd.DataFrame(fvalues, index=categorial_features_all, columns=numeric_features)
pvalues_df = pd.DataFrame(pvalues, index=categorial_features_all, columns=numeric_features)
```

Построим тепловые карты с полученными значениями F-статистики и значимости.

```python
plt.figure(figsize=(16, 16))

plt.subplot(121, title='F-score одностороннего теста ANOVA')
sns.heatmap(fvalues_df, robust=True, square=True, annot=True)
plt.xlabel('Количественные признаки')
plt.ylabel('Категориальные признаки')

plt.subplot(122, title='P-value одностороннего теста ANOVA')
sns.heatmap(pvalues_df, robust=True, square=True, annot=True, vmin=0, vmax=1)
plt.xlabel('Количественные признаки')
plt.ylabel('Категориальные признаки');
```

Там, где значение р-value больше 0,05, мы не можем отвергнуть нашу нулевую гипотезу и не можем принять нашу альтернативу. Отсюда следует, что столбец *'Gender'* статистически незначимо связан с признаками *'MonthlyCharges'*, *'TotalCharges'* и *'CustomerLifetime'*; признак *'InternetService'*, *'PaperlessBilling'*, *'SeniorCitizen'* - с признаком *'CustomerLifetime'*. ## Построение моделей

### Построение и отбор бейслайнов

* Создадим конвейер с предобработкой категориальных признаков методом OHE и моделью машинного обучения.
* Для задачи выберем четыре модели: логистическую регрессию, классификатор на методе опорных векторов, случайный лес и градиентный бустинг.
* Зададим необходимые значения для генерации случайных чисел и количество процессоров для вычисления, остальное оставим по умолчанию.
* Для надёжности результатов обучения проведём кросс-валидацию на обучающем множестве 5 раз.
* Оценивать качество будем по двум метрикам: *roc_auc* и *accuracy*.

Начнём с полной выборки без учёта несбалансированности классов.

```python
categories = [customer_full_df[col].unique().tolist() for col in categorial_features]

preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(categories=categories, drop='first'), categorial_features)],
    remainder='passthrough',
    n_jobs=-1
)

logit = LogisticRegression(random_state=RANDOM_STATE, n_jobs=-1)
svc = SVC(random_state=RANDOM_STATE)
rfc = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1)
lgbc = lgb.LGBMClassifier(random_state=RANDOM_STATE)

models_dict = {
    'Логистическая регрессия': logit,
    'Классификатор на опорных векторах': svc,
    'Случайный лес': rfc,
    'Градиентный бустинг': lgbc
}

pipes_dict = {
    desc: Pipeline([('preprocessor', preprocessor), ('classifier', model)])
    for desc, model in models_dict.items()
}

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

cv_scores_dict = {}
metrics_list = ['roc_auc', 'accuracy']

for desc, pipe in pipes_dict.items():
    cv_scores = cross_validate(
        estimator=pipe,
        X=features_train,
        y=target_train,
        scoring=metrics_list,
        cv=skf,
        return_train_score=True,
        n_jobs=-1
    )
    
    mean_cv_scores_dict = {}
    
    for stage in ['train_', 'test_']:
        for metric in metrics_list:
            mean_cv_scores_dict['mean_' + stage + metric] = cv_scores[stage + metric].mean()
    
    cv_scores_dict[desc] = mean_cv_scores_dict

total_cv_scores = [cv_scores_dict]
```

### Подбор гиперпараметров

Создадим пайплайн и необходимые компоненты для подбора гиперпараметров случайного леса по случайной сетке и обучим его.

```python
# Предопределим категории для трансформера
categories = [customer_full_df[col].unique().tolist() for col in categorial_features]
# Инициализируем трансформер
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(categories=categories, drop='first'), categorial_features)],
    remainder='passthrough',
    n_jobs=-1
)
# Инициализируем случайный лес
rfc = RandomForestClassifier(random_state=RANDOM_STATE, n_jobs=-1, class_weight='balanced')
# Инициализируем пайплайн
pipe = Pipeline([('preprocessor', preprocessor), ('classifier', rfc)])
# Инициализируем стратифицированное перемешивание
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
# Инициализируем сетку параметров
cv_params = {
    'classifier__n_estimators': range(10, 110, 5),
    'classifier__max_depth': range(5, 25, 2),
    'classifier__min_samples_leaf': range(1, 18, 2)
}

rfc_grid = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=cv_params,
    n_iter=50,
    scoring='roc_auc',
    n_jobs=-1,
    cv=skf,
    random_state=RANDOM_STATE,
    return_train_score=True
)

rfc_grid.fit(features_train, target_train)
```

Перейдём к градиентному бустингу.

```python
# Найдём оставшиеся категориальные признаки
new_categorial_features = list(set(categorial_features) - set(cols_to_drop))
# Предопределим категории для трансформера
categories = [customer_full_df[col].unique().tolist() for col in new_categorial_features]
# Инициализируем трансформер
preprocessor = ColumnTransformer(
    transformers=[('cat', OneHotEncoder(categories=categories, drop='first'), new_categorial_features)],
    remainder='passthrough',
    n_jobs=-1
)
# Инициализируем градиентный бустинг
lgbc = lgb.LGBMClassifier(random_state=RANDOM_STATE, is_unbalance=True)
# Инициализируем пайплайн
pipe = Pipeline([('preprocessor', preprocessor), ('classifier', lgbc)])
# Инициализируем стратифицированное перемешивание
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
# Инициализируем сетку параметров
cv_params = {
    'classifier__n_estimators': range(100, 600, 100),
    'classifier__max_depth': range(5, 13, 2),
    'classifier__num_leaves': range(30, 120, 10),
    'classifier__min_child_samples': range(20, 100, 10)
}

lgbc_grid = RandomizedSearchCV(
    estimator=pipe,
    param_distributions=cv_params,
    n_iter=50,
    scoring='roc_auc',
    n_jobs=-1,
    cv=skf,
    random_state=RANDOM_STATE,
    return_train_score=True
)

lgbc_grid.fit(features_after_corr_train, target_train)
```

В заключении построим график важности признаков. Так как перед обучением данные были кодированы методом OHE, то количество признаков будет больше, чем в тренировочной выборке.

```python
def add_value_labels(ax, spacing=5):
    """
    Add labels to the end of each bar in a bar chart.

    Arguments:
        ax (matplotlib.axes.Axes): The matplotlib object containing the axes
            of the plot to annotate.
        spacing (int): The distance between the labels and the bars.
    """
    rects = ax.patches
    
    # For each bar: Place a label
    for rect in rects:
        # Get X and Y placement of label from rect.
        x_value = rect.get_width()
        y_value = rect.get_y() + rect.get_height() / 2

        # Number of points between bar and label. Change to your liking.
        space = 5
        # Vertical alignment for positive values
        ha = 'left'

        # If value of bar is negative: Place label left of bar
        if x_value < 0:
            # Invert space to place label to the left
            space *= -1
            # Horizontally align label at right
            ha = 'right'

        # Use X value as label and format number with one decimal place
        label = "{:.3f}".format(x_value)

        # Create annotation
        ax.annotate(
            label,                      # Use `label` as label
            (x_value, y_value),         # Place label at end of the bar
            xytext=(space, 0),          # Horizontally shift label by `space`
            textcoords="offset points", # Interpret `xytext` as offset in points
            va='center',                # Vertically center label
            ha=ha                       # Horizontally align label differently for
        )                               # positive and negative values.


new_cols = []
for col in features_after_corr.columns:
    if col in new_categorial_features:
        new_cols.extend(features_after_corr[col].unique().tolist()[1:])
    else:
        new_cols.append(col)

importances = lgbc_grid.best_estimator_['classifier'].feature_importances_
importances = importances / importances.sum()

feature_importances = pd.DataFrame(
    {'importance': importances},
    index=new_cols
).sort_values('importance')

ax = feature_importances.plot(
    kind='barh',
    xlim=(0, feature_importances['importance'].max()*1.1),
    title='Важность признаков в построенной модели "градиентный бустинг"',
    figsize=(10, 8),
    legend=False
)

ax.set_xlabel('Важность')
ax.set_ylabel('Признак')
add_value_labels(ax)
```

График показывает, что наибольшее влияние на прогноз оказывают время жизни клиента (45.2% от всех признаков) и подключен ли интернет по оптоволокну (38.4% от всех признаков).
